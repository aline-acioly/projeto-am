# 02 - Demo dos baselines (LightGBM, XGBoost, CatBoost)

Roda os três baselines em um dataset pequeno do TabArena para verificar que o pipeline `split + fit + evaluate` funciona end-to-end. Se este notebook executa sem erro, o ambiente está pronto para o experimento completo.

```python
import sys
from pathlib import Path

sys.path.append(str(Path('..').resolve()))

from data.load_tabarena import load_task, RECOMMENDED_TASK_IDS
from src.models.baselines import BASELINE_FACTORIES
from src.pipeline.split import stratified_split
from src.pipeline.evaluate import fit_predict_evaluate
```

```python
ds = load_task(RECOMMENDED_TASK_IDS[0])
X_train, X_test, y_train, y_test = stratified_split(ds.X, ds.y, seed=42)
print(f'Dataset: {ds.name} (n={ds.n_samples}, p={ds.n_features}, k={ds.n_classes})')
```

```python
for name, factory in BASELINE_FACTORIES.items():
    estimator = factory(seed=42)
    metrics = fit_predict_evaluate(estimator, X_train, y_train, X_test, y_test)
    print(f'{name:>10s}: AUC={metrics.auc_ovo:.4f}, ACC={metrics.accuracy:.4f}, time={metrics.fit_time_s + metrics.predict_time_s:.1f}s')
```
