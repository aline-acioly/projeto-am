# 03 - Demo do modelo do grupo

Antes de rodar este notebook, edite `src/models/group_model.py` para retornar uma instância do modelo atribuído ao seu grupo. A lista de modelos atribuíveis está descrita no README. Veja os exemplos comentados no próprio arquivo.

```python
import sys
from pathlib import Path

sys.path.append(str(Path('..').resolve()))

from data.load_tabarena import load_task, RECOMMENDED_TASK_IDS
from src.models.group_model import build_group_model
from src.pipeline.split import stratified_split
from src.pipeline.evaluate import fit_predict_evaluate
```

```python
ds = load_task(RECOMMENDED_TASK_IDS[0])
X_train, X_test, y_train, y_test = stratified_split(ds.X, ds.y, seed=42)

model = build_group_model(seed=42)
metrics = fit_predict_evaluate(model, X_train, y_train, X_test, y_test)
print(f'AUC={metrics.auc_ovo:.4f}, ACC={metrics.accuracy:.4f}, time={metrics.fit_time_s + metrics.predict_time_s:.1f}s')
```
